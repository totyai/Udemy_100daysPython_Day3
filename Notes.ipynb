Write code to check if the inputted number is even or odd

```python
number = int(input("Add a whole number: "))
remainder = number % 2

if remainder == 0:
    print("You have choose an even number!")
else:
    print("You have choose an odd number!")
```

## Nested statements and elif statement

This means that we can check two statements within one another. It will evaluate from outter to inner.
And add with indentation the other statement.

Elif statement add one more or more conditions into the initial statements. There is no limit how much elif can be added

```python
print("Welcome to the rollercoaster!")
height = int(input("What is your height in cm? "))

if height >= 120:
    print("You can ride the rollercoaster")
    age = int(input("What is your age? "))
    if age <= 12:
        print("please pay $5.")
    elif age <= 18:
        print("Please pay $7.")
    else:
        print("Please pay $12.")
else:
    print("Sorry you have to grow taller before you can ride.")
```
