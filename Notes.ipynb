## Multiple ifs in succession

If there is an additional condition, that is not dependent of the other we can chain them outside of else statement

```python
print("Welcome to the rollercoaster!")
height = int(input("What is your height in cm? "))
bill = 0

if height >= 120:
    print("You can ride the rollercoaster")
    age = int(input("What is your age? "))
    if age <= 12:
        print("Child tickets are $5.")
        bill = 5
    elif age <= 18:
        print("Youth ticket are $7.")
        bill = 7
    else:
        print("Adult tickets are $12.")
        bill = 12
    
    #Chained if
    want_photo = input("Do you want to have a photo taken? Tpye y or n.")
    if want_photo == "y":
        #Add $3 to their bill
        bill += 3
    
    #This print will run even if they don't want the photo
    print(f"Your final bill is ${bill}")
else:
    print("Sorry you have to grow taller before you can ride.")
```
